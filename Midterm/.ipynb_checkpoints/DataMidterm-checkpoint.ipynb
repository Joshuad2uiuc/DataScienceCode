## Problem 1: Email Address Matching
** 4 points **

You are given a list of names and a list of email addresses.  How would you automatically assign the 'best' email from the list of emails to the corresponding name from the list of names?

You can find a list of names in `./names.txt` and the list of emails in `./emails.txt`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import requests
import re
import math
from matplotlib.ticker import MaxNLocator
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import nltk
import csv
from bs4 import BeautifulSoup 
import re 
import json
from nltk.corpus import stopwords
```

```python
# This probably won't work in non-*nix operating systems
!head names.txt
!echo
!head emails.txt
```

```python
'''
Loop through each name, look for the best email using regex. 
'''
def email_matcher(emails_file, names_file):
    matches = {}

    with open(names_file, 'r') as names:
        for i in names:
            with open(emails_file, 'r') as emails:
                first = i[:(i.index(' '))]
                init = i[0]
                last = i[::-1].strip()
                last = last[0:(last.index(' '))][::-1]
                pattern1 = (first + '.' + last).lower()
                pattern2 = (init + last).lower()
                for j in emails:
                    m1 = re.match(pattern1,j)
                    m2 = re.match(pattern2,j)
                    if m1:
                        matches[i] = j[:len(j)-1]
                    elif m2:
                        matches[i] = j[:len(j)-1]
                    else:
                        continue
    return matches
    pass

email_matcher('emails.txt', 'names.txt')
```

## Problem 2: San Francisco City Salaries
** 2 points **

In `Salaries.csv`, you'll find a list of pay data for the public employees of San Francisco City for the years 2011 - 2014

Write a function that will take this data and plot the median pay change between 2011 and 2014. You can limit your visualization to the top 25 most populous professions.

(Data courtesy of https://www.kaggle.com/kaggle/sf-salaries)

```python
!head Salaries.csv
```

```python
'''
Get the top 25 most populous jobs, get the median pay each year, 
find the change in median pay between the years,
plot it using a histogram, best way I could find it.
I am assuming that Police Officer and Police Officer 2 are of different rankings hence 
different pay.
'''

def median_sf_salaries_change(filename):
    salaries_df = pd.read_csv(filename, sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
    salaries_df['JobTitle'] = salaries_df.JobTitle.str.lower()
    for i in xrange(len(salaries_df.keys())):
        salaries_df.min().fillna(i)
    topJobs = np.array(salaries_df['JobTitle'].value_counts()[0:25].keys())
    yValues = [] 
    baseYear = 2011
    for i in topJobs:
        count = 0
        avg = 0
        last = 0
        for j in xrange(0,4):
            a = salaries_df[salaries_df.Year == str(baseYear+j)][salaries_df.JobTitle == i].TotalPay.median()
            if(math.isnan(a) != True and count == 0):
                count += 1
            elif(math.isnan(a) != True):
                avg += (last - a)
                count += 1
            else:
                continue
            last = a
        yValues.append(avg/count) 
    fig, ax = plt.subplots()

    xValues = [x for x in xrange(len(topJobs))]
    plt.rcdefaults()

    plt.xlabel('Change in median pay from 2011-2014')
    plt.title('Average Median Change in Pay for top 25 most populous San Fran professions')
    plt.bar(xValues, yValues, align='center')
    plt.xticks(xValues, list(topJobs), rotation='vertical')
    plt.show()
    pass

median_sf_salaries_change('Salaries.csv')
```

## Problem 3: Airline Tweets
** 3 points **

Given a dataset of Tweets regarding different airlines (`Tweets.csv`), write a function that returns an ordered list of the most serious complaints against the specifiec airline.

**Note:** The sentiment analysis has already been performed, and you are given a sentiment score, complaint label (`negativearesas`), and complaint label confidence markers for each tweet. For maximum accuracy, you may wish to utilize all three of these values in your analysis.

(Data courtesy of https://www.kaggle.com/crowdflower/twitter-airline-sentiment)

```python
!head Tweets.csv
```

```python
'''
I sorted the dataframe by top retweets descending to lowest. The more retweets them more
serious the complaint because it spreads to more people. If the confidence, is full for negative
and the analysis, if it is negative, and if the tweet has at least one retweet, I considered it
to be a serious complaint
'''
def airline_complaints(filename, airline):
    tweets_df = pd.read_csv(filename, sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
    tweets_df = tweets_df.sort('retweet_count',ascending=False)
    data = tweets_df[tweets_df.airline == airline][tweets_df.negativereason_confidence == '1.0'][tweets_df.airline_sentiment_confidence == '1.0'][tweets_df.airline_sentiment == 'negative'][tweets_df.retweet_count != '0'].text
    return list(data)
    pass

airline_complaints('Tweets.csv', 'American')
```

## Problem 4: City Name Cleansing
** 3 points**

Suppose you have a very large list of the names of all large cities in the world. (Number of records ~= 100,000) Many of these cities will contain common overlapping words like “Mobile”, “Rugby” or “Salmon”. Your task is to automatically (and hopefully quickly) filter out as many of these 'common' words as possible.

Unfortunately, we don't have a clean list of cities, but rather a \*.csv file that contains a bunch of other information. You can find this file in `worldcities.csv`

(Data courtesy of https://www.maxmind.com/en/free-world-cities-database)

```python
!tail worldcities.csv
```

```python
'''
Filtered out the top 500 most common words in the file
'''

def city_name_cleanser(filename):
    cities_df = pd.read_csv(filename, sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
    fdist = FreqDist()
    values = []
    for x in xrange(len(stopwords.words('english'))):
        values.append(x)
    stopDict = dict(zip(stopwords.words('english'), values))
    for i in cities_df.City:
        for j in word_tokenize(i):
            if(j not in stopDict and len(j)>2):
                fdist[j]+=1
    topCommon = {}
    cities = fdist.keys()
    count = fdist.values()
    for i in xrange(500):
            index = count.index(max(count))
            topCommon[cities[index]] = count[index]
            del cities[index]
            del count[index]
    count = 0
    for city in cities_df.City:
        changed = word_tokenize(city)
        for i in changed:
            if(i in topCommon):
                changed.remove(i)
                if len(changed) != 0:
                    cities_df.set_value(count,'City',' '.join(changed))                    
        count+=1

    return cities_df
    pass

city_name_cleanser('worldcities.csv')
```

** Note: ** You're encouraged to separate your code into function(s) that scrape data from IMDB, and function(s) that perform the crawling logic.

```python
'''
This one was self explanatory, output was intended and it worked for multiple movie pages for me.
'''
def getUrls(seed_url):
    imdb_url = seed_url
    imdb_links = {} 
    links_to_scrape = [imdb_url]
    while links_to_scrape and len(imdb_links) < 100:
        scrape_page = links_to_scrape.pop()
        soup = BeautifulSoup(requests.get(scrape_page).text)
        for link in soup.find_all('a'):
            href = link.get('href')
            if not href or not href.startswith('/title/tt'):
                continue
            if not href in imdb_links:
                imdb_links[href] = [scrape_page]
                if href not in links_to_scrape:
                    links_to_scrape.append(href)
            else:
                imdb_links[href].append(scrape_page)
    movieList = []
    for i in imdb_links:
        if i.index('?') == 17 and i.index('r') ==18:
            movieList.append('http://www.imdb.com' + i)
    return movieList
    pass

def imdbScraper(seed_url, out_file):
    movieUrls = getUrls(seed_url)
    movieJson = []
    
    for i in xrange(len(movieUrls)):
        soup = BeautifulSoup(requests.get(movieUrls[i]).text)
        title = soup.find("meta",  property="og:title")['content']
        rating = soup.find("span", class_="rating").contents[0]
        duration = str(soup.find("time", itemprop="duration").contents[0]).strip()
        date = soup.find(title="See more release dates").contents[0]
        for i in soup.find_all('h4', class_="inline"):
            if i.contents == [u'Budget:']:
                budget = i.next_sibling.strip()
        movieJson.append({
            'Title': title,
            'Rating out of 10': rating,
            'Duration': duration,
            'Date': date
            })
    with open(out_file, 'w') as outfile:
        json.dump(movieJson, outfile)
    return 'Completed'
    pass
            

imdbScraper('http://www.imdb.com/title/tt0068646/?ref_=tt_rec_tti', 'movies.json')
```
